````python
car_sales_with_missing_data=pd.read_csv('./car-sales-extended-missing-data.csv')
````

 *  we cant predict the values or even fit these values into the model

>>>>>>> Method 1: Fill missing data with pandas

````python


#Fill the 'make' Column
car_sales_with_missing_data["Make"].fillna('missing',inplace=True)


#Fill the 'Colour' Column
car_sales_with_missing_data["Colour"].fillna('missing',inplace=True)


#Fill the 'Odometer' column with mean value
car_sales_with_missing_data['Odometer (KM)'].fillna(car_sales_with_missing_data['Odometer (KM)'].mean(),inplace=True)


#Fill the 'Doors' column with mannual wit i.e avg of 4,5,3 is 4 so 
car_sales_with_missing_data['Doors'].fillna(4,inplace=True)
````

````python
#Now you can  see we have 0 missing values except the Price Column
car_sales_with_missing_data.isna().sum()

#We cant fill it up with anything cuz its whats meant to be predicted so we gonna simply remove it 
````

````python
#Removing the records havivng price missing values
car_sales_with_missing_data.dropna(inplace=True)

````

````python
#now train model again 

X=car_sales_with_missing_data.drop("Price",axis=1)

Y=car_sales_with_missing_data['Price']
````

>>>>>> Method 2: Fill missing values with scikit-Learn

````python
car_missing_data=pd.read_csv('./car-sales-extended-missing-data.csv')
car_missing_data.isna().sum()
````

````python
#First remove the records having the Price values missing as Y is not meant to be missing
car_missing_data.dropna(subset=["Price"],inplace=True)
````

````python
car_missing_data.isna().sum()
````

````python
#Now Split the data to  features and labels 

X=car_missing_data.drop("Price",axis=1)
Y=car_missing_data["Price"]
````

````python
X
````

````python
Y
````

````python
#Fill the data with sklearn 

from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

#Ab voi same .. fill the  Categorical data wtih 'missing' and the numerical data with 
#mean 


cat_imputer = SimpleImputer(strategy="constant", fill_value="missing")
door_imputer= SimpleImputer(strategy="constant", fill_value=4)
num_imputer= SimpleImputer(strategy="mean")



#Define Columns 
categorical_features = ['Make',"Colour"]
door_features = ["Doors"]
num_features=["Odometer (KM)"]


#Create an imputer (Something that fills missing data)

imputer=ColumnTransformer([
    ("cat_imputer",cat_imputer,categorical_features),
    ("door_imputer",door_imputer,door_features),
    ("num_imputer",num_imputer,num_features),
])


#Transform the data

filled_X=imputer.fit_transform(X)
````

````python
#give the column names to the matrix of data we got 
car_sales_filled=pd.DataFrame(filled_X,columns=['Make','Colour','Doors','Price'])
````
