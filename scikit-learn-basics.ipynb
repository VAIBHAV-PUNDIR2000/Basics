### <center>1. Getting the data ready</center>

````python
import pandas as pd
import numpy as np
heart_disease=pd.read_csv("./heart-disease.csv")
heart_disease
````

````python
#So target hai ki target predict krenge baakin features se

#Create X: Feature matrix: 

X=heart_disease.drop("target",axis=1)
X
````

````python
#Create Y: Labels:
Y=heart_disease["target"]
Y
````

### <center>2.Choose The right model and <strong>Hyperparameters</strong>:
</center>
<em

>

><strong>Hyperparameters</strong> are like the radio dials that in radio that helps us to fine tunee the signals and get crisp and perfect audio in channel

````python
from sklearn.ensemble import RandomForestClassifier
new_clf=RandomForestClassifier()        #a classification machine learning model
````

````python
new_clf.get_params()                 #Keep the default Hyperparameters
````

### <center>3. Split and Fit the model into the training algo </center>

````python
from sklearn.model_selection import train_test_split  #for splitting data

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
````

Now lets see what distribution has been done with this function
***

#### <center>Training Data</center>

````python
print(len(X_train))   #Notice the 212 records without target
X_train 
````

````python
print(len(Y_train))     #Notice the 212 records only consisting target 
Y_train 
````

#### <center>Testing Data</center>

````python
print(len(X_test))     #Notice the 91 records without target
X_test                 
````

````python
print(len(Y_test))     #Notice the 91 records with target only 
Y_test  
````

#### <center>Fitting the data </center>

````python
new_clf.fit(X_train,Y_train)
````

#### <center>Prediction </center>

````python
#store your predictions into a series
predictions=new_clf.predict(X_test)
predictions
````

## <center  class="alert alert-block alert-info"> 4. Evaluate the Model based on the predictons made from learning data and test data</center>

````python
new_clf.score(X_train,Y_train)
````

````python
new_clf.score(X_test,Y_test)
````

````python
#Other Matrices or methods to evaluate we can use to meaure accuracy 

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

````

````python
#Classification_report

print(classification_report(Y_test,predictions))
````

````python
#Confusion matrix

print(confusion_matrix(Y_test, predictions))
````
