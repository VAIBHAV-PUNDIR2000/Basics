### The 3 methods of using Matplotlib to plot

```python
#1st method
fig=plt.figure()            #Creates a figure without Axis
ax= fig.add_subplot()       #getting access to the axes
plt.show()

#same as 
plt.plot()
```

```python
#2nd method
fig= plt.figure() #creates a figure
ax=fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()  #same as what we plotted above just thoda lamba hai 

```

```python
#3rd method (Recommended)

fig,ax=plt.subplots()           #every call to this resets the data previously saved
ax.plot(x,[10.20,23,45,12])    # add the data to axes
type(fig),type(ax)  #output shows why object oriented 
# fig.show()
```

## Making figures from numpy arrays

#### We want :
* Line Plot
* Scatter Plot
* Bar plot
* Histogram
* Subplots


```python
import numpy as np
```

```python
#create the data 

x=np.linspace(1,10,100)
x
```

```python
#plot the data and create a line plot 
#default is the line plot

fig, ax= plt.subplots()
ax.plot(x,x**2)


```

```python
#use same data to make scatter plots

fig,ax=plt.subplots()
ax.scatter(x=x,y=np.exp(x))
```

```python
#Another scatter plot 

fig,ax= plt.subplots()
ax.scatter(x=x,y=np.sin(x))
```

```python
#Another bar Plot from dictionary


#data formatting 

butter_prices={"Almond":10,"Peanut":8,"Cashew":12};


#gathering the plot objects

fig,ax= plt.subplots();


#passing data to plot axes
ax.bar(butter_prices.keys(),butter_prices.values());

#decoration 
ax.set(title="Prices",xlabel="Different butter",ylabel="Amount in $"
);
```

```python
#Horizontal bar graph



fig,ax= plt.subplots()
ax.barh(list(butter_prices.keys()),list(butter_prices.values()),
height=0.5,
color="black")
```

```python
#Histograms

x=np.random.randn(1000)
fig, ax= plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
#1. subplot option 1

fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,

ncols=2,figsize=(10,5))

ax1.plot(x,x/2)
ax2.scatter(np.random.randint(0,10000,100),np.random.randint(0,1000,100))
ax3.barh(list(butter_prices.keys()),list(butter_prices.values()))
ax4.hist(np.random.randint(0,100,100))
```

```python
# Subplots options 2
fig, ax =plt.subplots(nrows=2,ncols=2)
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.randint(0,10000,100),np.random.randint(0,1000,100))
ax[1,0].barh(list(butter_prices.keys()),list(butter_prices.values()))
ax[1,1].hist(np.random.randint(0,100,100));
```

### which one to use ?  (pyplot vs matplotlib OO method?)

* When plotting something quicky its okay to use pyplot method

* when plotting something advanaced use OO mthod

```python
#using ddata from csv
over50=ds[ds["age"]>50]
over50.head()
```

```python
#pyplot method
over50.plot(kind="scatter",x="age",y="chol",c="target");
```

```python
# no oo method 
fix , ax = plt .subplots(figsize=(10,6))
over50.plot(kind="scatter",x="age",
y="chol",
c="target",
ax=ax
);
```

```python
##oo method from scratch




fig,ax=plt.subplots();


scatter=ax.scatter(x=over50["age"]
,y=over50["chol"],c=over50["target"]);


#customise the plot
ax.set(title="Heart disease and cholestrol levels",
xlabel="Ages",ylabel="Cholestrol levels",);


#add a legend
ax.legend(*scatter.legend_elements(),title="Target")

#add a line to the plot
# ax.plot(over50["thalach"])
ax.axhline(y=np.mean(over50["chol"])
,linestyle="--"
);
```

```python
#Subplot of chol age thalach 


fig,ax= plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)


#add data to x=0
ax[0].scatter(x=over50["age"],
y=over50["chol"],
c=over50["target"],
)

#costumization  ax[0]
ax[0].set(title="cholestrol level per age group ",
# xlabel="ages"

ylabel="cholestrol level"
);

#adding legends()

ax[0].legend(*scatter.legend_elements(),title="target")


#adding a mean line 

ax[0].axhline(y=over50["chol"].mean())


#add data to x=1
ax[1].scatter(x=over50["age"],
y=over50["thalach"],
c=over50["target"],
)

#costumization  ax1]
ax[1].set(title="thalach level per age group ",
xlabel="ages",

ylabel="thalach level"
);

#adding legends()

ax[1].legend(*scatter.legend_elements(),title="target")


#adding a mean line 

ax[1].axhline(y=over50["thalach"].mean())

fig.suptitle("Heart disease Analysis")
```
